# 프로젝트 목표

- 비즈니스 이해를 통한 머신러닝 모델 활용 계획 수립
- 머신러닝 모델 구축을 위한 데이터 셋 준비 및 전처리
- 머신러닝 모델과 딥러닝 모델 학습 및 평가
- 평가를 통한 성능이 좋은 최적의 모델 설정 및 배포

# 세부 수행 절차 및 내용

- 프로젝트와 관련된 비즈니스 목표 이해
- 데이터 탐색적 분석 및 데이터 전처리
- 학습용 데이터셋과 테스트 데이터셋 분리
- 활용할 머신러닝/딥러닝 모델들 후보 선정하고 모델링 진행
- 다양한 모델링 결과 측정 지표로 측정하고 최적의 모델 선정
- 최적의 모델 배포 및 테스트

# 주요 산출물

- 인공지능 데이터 전처리 결과서
- 인공지능 학습 결과서
- 학습 된 인공지능 모델

# 프로젝트 수행 상세 가이드

- 문제 정의 (비즈니스 이해)
- 데이터 수집 / 확인
- EDA (탐색적 데이터 분석)
- 데이터 전처리
- Feature Engineering
- 데이터 분리 (train/test)
- 모델링 (여러 모델 비교)
- 성능 평가 및 모델 선택
- 결과 해석
    - 모델의 성능 뿐 아니라 주요 feature가 이탈에 어떤 영향을 미치는지 해석 포함
- 서비스 형태로 배포 (Streamlit)
- 모델은 전처리와 함께 Pipeline 형태로 구성하고 파일로 저장하여 재사용 가능하게 함

# Github 디렉토리 구조 예시

```python
project-root/
│
├── README.md              # 프로젝트 전체 설명 (문제 정의, 데이터, 모델, 결과, 실행 방법)
├── requirements.txt       # 실행에 필요한 라이브러리 목록
├── .gitignore             # Git에 업로드하지 않을 파일/폴더 설정
│
├── data/
│   ├── raw/               # 원본 데이터 (수정 금지)
│   ├── processed/         # 전처리 완료 데이터 (모델 입력용)
│
├── notebooks/
│   ├── 01_eda.ipynb        # 데이터 탐색 및 시각화
│   ├── 02_preprocessing.ipynb # 전처리 과정 실험 및 확인
│   ├── 03_modeling.ipynb   # 모델 학습 및 성능 비교
│
├── src/
│   ├── preprocessing.py   # 전처리 로직 함수화 (재사용 목적)
│   ├── train.py           # 모델 학습 코드 (실행용)
│   ├── predict.py         # 학습된 모델로 예측 수행
│
├── models/                # 학습 완료된 모델 파일
│
├── app/
│   ├── app.py             # Streamlit 기반 예측 서비스 실행 파일
│
└── reports/
    ├── eda_report.md      # 데이터 분석 결과 정리 문서
    ├── model_report.md    # 모델 성능 비교 및 선택 이유 정리
```

# README 구성 예시

- 프로젝트 개요
    - 문제 정의

- 사용 기술
- 데이터 설명
    - 데이터 출처
    - 주요 컬럼 설명
    - 타겟 변수 정의

    - 핵심 인사이트

    - 결측치 처리
    - 이상치 처리
    - 인코딩 / 스케일링
- 모델링
    - 사용 모델 목록
    - 선택 이유
- 성능 비교
    - 표 형태로 정리
- 최종 모델 선택 이유
    - 단순 accuracy 말고 근거 필요
    - feature 중요도 또는 해석 결과 포함
- 배포 방법
    - 실행 방법 (streamlit run app/app.py)
- 결과 화면
    - 캡처 이미지
- 회고

# 주요 산출물 상세 기준

### 1) 데이터 전처리 결과서 (eda_report.md)

- 타겟 변수 분포 (불균형 여부)
- 주요 feature와 이탈 관계
- 이상치 존재 여부
- 결측치 처리 방식

그래프와 해석을 덧붙인 구조로 작성

### 2) 모델 학습 결과서 (model_report.md)

- 모델별 성능 비교
- Confusion Matrix 해석
- 어떤 상황에서 모델이 틀리는지 분석

# 평가 기준

- 데이터 탐색 및 전처리가 논리적으로 수행되었는가?
    - EDA, 결측치/이상치 처리, 인코딩/스케일링 포함
- 다양한 모델을 비교하고, 최종 모델 선택에 대한 근거가 명확한가?
    - 단순 성능이 아닌 이유 설명 포함
- 성능 평가 지표를 적절히 활용하고, 결과를 해석할 수 있는가?
    - Confusion Matrix, Precision/Recall 등
- 모델 성능 개선을 위한 시도가 있었는가?
    - 튜닝, 과적합 방지 등
- 코드 및 결과가 일관성 있게 정리되어 있으며 재현 가능한가?
- 사용자 입장에서 사용할 수 있는 형태로 서비스가 구현되었는가?
    - Streamlit, 입력 → 예측 흐름, 실행 가능 여부
